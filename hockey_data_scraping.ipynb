HOCKEY DATA SCRAPING

```python
#Importing the Libraries
```

```python
from bs4 import BeautifulSoup
import requests
```

```python
#Url of the website which has to be scraped
```

```python
url = 'https://www.scrapethissite.com/pages/forms/?page_num=6&per_page=100'
```

```python
page = requests.get(url)
```

```python
soup = BeautifulSoup(page.text , 'html')
print(soup)
```

```python
#Finding the class 'table' in the soup (website-data)
```

```python
hockey_data = soup.find_all('table')
print(hockey_data)
```

```python
#Finding the table headers using the 'th'
```

```python
hockey_titles = soup.find_all('th')
```

```python
hockey_titles
```

```python
#Formatting the text using .text().strip() functions
```

```python
hockey_table_titles = [title.text.strip() for title in hockey_titles]

print(hockey_table_titles)
```

```python
#Importing pandas library for storing the data in dataframe
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = hockey_table_titles)

df
```

```python
#Iterarting through each row and inserting the data in the dataframe
```

```python
row_data = soup.find_all('tr')
```

```python
for row in row_data[1:]:
    column_data = row.find_all('td')
    individual_column_data = [data.text.strip() for data in column_data]
    Length = len(df)
    df.loc[Length] = individual_column_data
```
